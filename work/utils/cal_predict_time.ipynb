```python
from keras.models import load_model
import os
import cv2
import time
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
base_path = 'C:/Users/user/OneDrive/ /Server/model/model_result'
deep_path = ['alexnet/alexnet.h5',
             'googlenet/googlenet.h5',
             'pdprog/pdprog.h5',
             'resnet_v2/resnet_v2.h5',
             'restnet101/resnet101.h5',
             'vgg16/vgg16.h5',
             'vgg19/vgg19.h5'
             ]
def image_preprocessing(image):
    image = np.array(image)
    image_resize = cv2.resize(image, dsize = (400, 400))
    image_hsv = cv2.cvtColor(image_resize, cv2.COLOR_RGB2HSV)
    image_hsv = image_hsv.astype(np.float)

    image_scaled = image_hsv.copy()
    image_scaled[:,:,0] = image_hsv[:,:,0] / 180.
    image_scaled[:,:,1] = image_hsv[:,:,1] / 255.
    image_scaled[:,:,2] = image_hsv[:,:,2] / 255.
    return image_scaled

test_gen = ImageDataGenerator(preprocessing_function=image_preprocessing)
test_flow_gen = test_gen.flow_from_directory(
    directory=f'./model_images/D012/train', color_mode = 'rgb',
    target_size=(400, 400), class_mode='categorical', batch_size=32, shuffle=False)
for i in deep_path:
    model = load_model(f'{base_path}/{i}')
    st = time.time()
    print(i)
    model.predict_generator(test_flow_gen)
    result_time = time.time() -st
    print(result_time)
    img_count = 628
    print(result_time/img_count)
    del model
```
